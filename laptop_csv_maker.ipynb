```python
import pandas as pd
import requests

from bs4 import BeautifulSoup # used for scraping
```

```python
with open("laptop.html", "r", encoding="utf-8") as f:
    html_content = f.read()
```

```python
# print(html_content[:500])  # print first 500 chars for preview
```

```python
soup=BeautifulSoup(html_content, 'lxml')
```

```python
# soup
```

```python
laptop_container=soup.findAll('div',class_='sm-product has-tag has-features has-actions')
```

```python
len(laptop_container)
```

```python
img = []
model_name = []
price = []
rating = []
specs_score = []
generation = []
core_and_thread = []
ram = []
rom = []
graphic_card = []
size_and_resolution = []
os = []
warranty = []

for i in laptop_container:
    # image
    try:
        img.append(i.find('div', class_='sm-img-wrap').find('img')['src'])
    except:
        img.append("Not Known Image")

    # model name
    try:
        model_name.append(i.find('div', class_='sm-img-wrap').find('img')['alt'])
    except:
        model_name.append("Not Known Model")

    # price
    try:
        price.append(i.find('span', class_='price').text)
    except:
        price.append("Not Known Price")

    # rating
    try:
        rating_str = i.find('span', class_='sm-rating')['style']  # "--rating: 4.05;"
        rating.append(float(rating_str.split(":")[1].replace(";", "").strip()))
    except:
        rating.append("Not Known Rating")

    # specs score (rank-1, 2 or 3)
    score = None
    try:
        rank1 = i.find("div", class_="score rank-1-bg")
        rank2 = i.find("div", class_="score rank-2-bg")
        rank3 = i.find("div", class_="score rank-3-bg")
        if rank1:
            score = rank1.text.strip()
        elif rank2:
            score = rank2.text.strip()
        elif rank3:
            score = rank3.text.strip()
    except:
        score = "Not Known Specs Score"
    specs_score.append(score)

    # features list
    try:
        features = i.find('ul', class_="sm-feat specs").find_all('li')
    except:
        features = []

    try:
        generation.append(features[0].text)
    except:
        generation.append("Not Known Generation")

    try:
        core_and_thread.append(features[1].text)
    except:
        core_and_thread.append("Not Known Core/Thread")

    try:
        ram.append(features[2].text)
    except:
        ram.append("Not Known RAM")

    try:
        rom.append(features[3].text)
    except:
        rom.append("Not Known ROM")

    try:
        graphic_card.append(features[4].text)
    except:
        graphic_card.append("Not Known Graphic Card")

    try:
        size_and_resolution.append(features[5].text)
    except:
        size_and_resolution.append("Not Known Size/Resolution")

    try:
        os.append(features[6].text)
    except:
        os.append("Not Known OS")

    try:
        warranty.append(features[7].text)
    except:
        warranty.append("Not Known Warranty")
```

```python
# dictionary with all lists
d = {
    'img': img,
    'model_name': model_name,
    'price': price,
    'rating': rating,
    'specs_score': specs_score,
    'generation': generation,
    'core_and_thread': core_and_thread,
    'ram': ram,
    'rom': rom,
    'graphic_card': graphic_card,
    'size_and_resolution': size_and_resolution,
    'os': os,
    'warranty': warranty
}

df=pd.DataFrame(d)
df.head()
```

```python
df.to_csv('laptop.csv',index=False)
```
