#### Cleaning

Order of Cleaning: It is crucial to follow a specific order when cleaning data for 
optimal results. This order is based on observations by experienced analysts: 
1. Completeness Issues: Address all missing data first. 
2. Messy Data (Structural Issues): Fix structural problems (e.g., un-tidying 
data by re-shaping, merging tables). 
3. Validity Issues: Correct data that is out of bounds or in an invalid format. 
4. Accuracy Issues: Fix incorrect or illogical data values. 
5. Consistency Issues: Standardise data representation.

```python
import re

mapping = {
    'single': 1, 'dual': 2, 'triple': 3, 'quad': 4,
    'penta': 5, 'hexa': 6, 'octa': 8, 'deca': 10
}

def get_core_number(value):
    val = str(value).strip().lower()
    
    # 1. Check text mapping first (quad → 4, octa → 8, etc.)
    for word, num in mapping.items():
        if word in val:
            return num
    
    # 2. Extract numeric part (e.g., "10 Cores", "24 Cores")
    digits = re.search(r'(\d+)', val)
    if digits:
        return int(digits.group(1))
    
    
    return None


copy_df['Core']=copy_df['Core'].apply(get_core_number)
```

```python
# copy_df[copy_df['Core'].isnull()]
```

```python
def get_thread_number(value):
    val = str(value).strip()
    
    # Extract digits safely
    match = pd.Series([val]).str.extract(r'(\d+)')[0]
    if not match.empty and pd.notna(match[0]):
        return int(match[0])
    
    return None

copy_df['Thread']=copy_df['Thread'].apply(get_thread_number)
```

```python
import re

def extract_gpu_info(value):
    
    val = str(value).replace("\u2009", " ").replace("\u200e", "").strip()

    # Brand
    if "NVIDIA" in val.upper():
        brand = "NVIDIA"
    elif "AMD" in val.upper():
        brand = "AMD"
    elif "INTEL" in val.upper():
        brand = "Intel"
    elif "APPLE" in val.upper():
        brand = "Apple"
    elif "QUALCOMM" in val.upper():
        brand = "Qualcomm"
    elif "ARM" in val.upper():
        brand = "ARM"
    elif "MEDIATEK" in val.upper():
        brand = "MediaTek"
    else:
        brand = "Unknown"

    # VRAM (like 2GB, 4GB, 8GB, 16GB, 24GB…)
    vram_match = re.search(r"(\d+)\s*GB", val.upper())
    vram = vram_match.group(1) + " GB" if vram_match else "Shared/Unknown"

    # Model / Series
    # Remove VRAM info before extracting
    clean_val = re.sub(r"\d+\s*GB", "", val, flags=re.IGNORECASE).strip()

    if brand != "Unknown":
        model = clean_val.replace(brand, "").strip()
    else:
        model = clean_val

    # Integrated vs Dedicated
    if "INTEGRATED" in val.upper() or brand in ["Intel", "Apple", "Qualcomm", "ARM", "MediaTek"]:
        gpu_type = "Integrated"
    else:
        gpu_type = "Dedicated"

    return pd.Series([brand, model, vram, gpu_type])
```
